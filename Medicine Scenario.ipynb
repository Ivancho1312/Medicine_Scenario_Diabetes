### 3. Find the patient which currently has the highest glucose level. Sketch (on a plot of your choice) all of the patients glucose levels and describe the plot. Draw a conclusion on how the data behaves.

Пронаоѓање на пациентот со моментално највисоко ниво на глукоза:

```python

max_glucose_pacient = df_filled[df_filled['Моментална крвна глукоза'] == df_filled['Моментална крвна глукоза'].max()]


tip_na_tretman = max_glucose_pacient['Тип на третман'].values[0]
if tip_na_tretman == 1.0:
    tip_na_tretman_text = 'Инсулин'
elif tip_na_tretman == 0.0:
    tip_na_tretman_text = 'Таблети'
else:
    tip_na_tretman_text = 'Без третман' 


dijabeticna_retinopatija = max_glucose_pacient['Дијабетична ретинопатија'].values[0]
if dijabeticna_retinopatija == 1.0:
    dijabeticna_retinopatija_text = 'Има'
elif dijabeticna_retinopatija == 0.0:
    dijabeticna_retinopatija_text = 'Нема'
else:
    dijabeticna_retinopatija_text = 'Непознато'
```

Анализирање на податоците за донесување на основен збирен заклучок за состојбата на пациентите и начините на третман од повеќе аспекти:

```python
avg_glucose_level = df_filled['Моментална крвна глукоза'].mean()
higher_than_average = df_filled[df_filled['Моментална крвна глукоза'] > avg_glucose_level]
count_higher_than_average = len(higher_than_average)

normal_range_min = 4.44
normal_range_max = 6.66
above_normal_range = df_filled[(df_filled['Моментална крвна глукоза'] > normal_range_max) | (df_filled['Моментална крвна глукоза'] < normal_range_min)]
count_above_normal_range = len(above_normal_range)

insulin_avg_glucose = df_filled[df_filled['Тип на третман'] == 1.0]['Моментална крвна глукоза'].mean()
tablets_avg_glucose = df_filled[df_filled['Тип на третман'] == 0.0]['Моментална крвна глукоза'].mean()
percentage_difference = ((insulin_avg_glucose - tablets_avg_glucose) / tablets_avg_glucose) * 100

comparison = ''
if percentage_difference > 0:
    comparison = f'Пациентите на инсулин имаат просечно ниво на глукоза повисоко од пациентите на таблети за {abs(percentage_difference):.2f}%.'
elif percentage_difference < 0:
    comparison = f'Пациентите на таблети имаат просечно ниво на глукоза повисоко од пациентите на инсулин за {abs(percentage_difference):.2f}%.'
else:
    comparison = 'Пациентите на инсулин и таблети имаат исто просечно ниво на глукоза.'
    
```

```python
# Печатење на резултатите и визуализација

print('Пациент со највисоко ниво на глукоза:\n')
print('ID:', max_glucose_pacient['ID'].values[0])
print('Времетраење на дијабетес (во год.):', 
      int(max_glucose_pacient['Времетраење на дијабетес (во год.)'].values[0]), 'год.')
print('Ниво на HbA1C во последните 3 месеци:', 
      max_glucose_pacient['Ниво на HbA1C во последните 3 месеци'].values[0])
print('Тип на третман:', tip_na_tretman_text)
print('Моментална крвна глукоза:', 
      max_glucose_pacient['Моментална крвна глукоза'].values[0])
print('Дијабетична ретинопатија:', dijabeticna_retinopatija_text)

print('-------------------------------------------------------')

print('\nHормални вредности на глукоза во крвта кај човекот се од 4.44 до 6.66 mmol/l.')
print('Просечно ниво на глукоза кај пациентите:', avg_glucose_level)
print('Број на пациенти со ниво на глукоза поголемо од просекот:', count_higher_than_average)
print('Број на пациенти со ниво на глукоза поголемо од нормалниот опсег:', count_above_normal_range)
print('Просечно ниво на глукоза за пациенти на инсулин:', insulin_avg_glucose)
print('Просечно ниво на глукоза за пациенти на таблети:', tablets_avg_glucose)
print(comparison)


plt.figure(figsize=(18, 8))
plt.bar(df['ID'], df_filled['Моментална крвна глукоза'])
plt.axhline(y=df_filled['Моментална крвна глукоза'].mean(), 
            color='red', linestyle='--', 
            label='Просечно ниво на глукоза')
plt.xlabel('Пациент ID')
plt.ylabel('Ниво на глукоза во крвта')
plt.title('Нивоа на глукоза во крвта за сите пациенти')
plt.legend()
plt.show()
```

### 5. Find the average duration of the disease for patients that are on insulin and patients that are on tablets. Compare the results. Which method that the patients use shows better results?

```python
insulin_avg_duration = df_filled[df_filled['Тип на третман'] == 1]['Времетраење на дијабетес (во год.)'].mean()
tablets_avg_duration = df_filled[df_filled['Тип на третман'] == 0]['Времетраење на дијабетес (во год.)'].mean()

days_in_year = 365
days_in_month = 30

insulin_avg_duration_days = int(round(insulin_avg_duration * days_in_year))
tablets_avg_duration_days = int(round(tablets_avg_duration * days_in_year))

insulin_years, insulin_days = divmod(insulin_avg_duration_days, days_in_year)
insulin_months, insulin_days = divmod(insulin_days, days_in_month)

tablets_years, tablets_days = divmod(tablets_avg_duration_days, days_in_year)
tablets_months, tablets_days = divmod(tablets_days, days_in_month)

print(f'Просечно времетраење на дијабетес кај пациенти на инсулин: {insulin_years} години, {insulin_months} месеци, {insulin_days} дена')
print(f'Просечно времетраење на дијабетес кај пациенти на таблети: {tablets_years} години, {tablets_months} месеци, {tablets_days} дена')
```

За да утврдиме кој метод дава подобри резултати потребни ни се повеќе информации. 
На пример:

Од друга страна пак, доколку болеста по пократкиот период на лечење е излечена, би значело подобар и поефективен тип на лечење на болеста. 
2. Во кој период од животот пациентот ја добил болеста? 
Не е исто да се анализираат и споредуваат ефектите на лековите кај пациенти со различна возраст. На пример, доколку пациент кај кој болеста настапила на 8 годишна возраст и се лечи последните 20 години со таблети, не можеме да го споредиме со пациент кај кој болеста настапила на 60 годишна возраст и се лечи со инсулин веќе 20 години. ### 6. A normal hemoglobin A1C glucose level is below 5.7. Divide the data in 2 groups (patients that have normal A1C level and patients that have elevated A1C level), then sketch the data using a plot of your choice dependent on the length of the disease. Distinguishing both groups.

```python
# Споредба помеѓу пациенти со нормално ниво и високо ниво на HbA1C

normal_hba1c = df_filled[df_filled['Ниво на HbA1C во последните 3 месеци'] < 5.7]
elevated_hba1c = df_filled[df_filled['Ниво на HbA1C во последните 3 месеци'] >= 5.7]

plt.figure(figsize=(16, 9))
plt.scatter(normal_hba1c['Времетраење на дијабетес (во год.)'], 
            normal_hba1c['Ниво на HbA1C во последните 3 месеци'], 
            label='Нормално ниво на HbA1C', 
            color='skyblue',
            alpha=0.7,
            s=50)
plt.scatter(elevated_hba1c['Времетраење на дијабетес (во год.)'], 
            elevated_hba1c['Ниво на HbA1C во последните 3 месеци'], 
            label='Покачено ниво на HbA1C', 
            color='salmon', 
            alpha=0.7, 
            s=50)

plt.xlabel('Времетраење на дијабетес (во год.)')
plt.ylabel('Ниво на HbA1C во последните 3 месеци')
plt.title('Ниво на HbA1C според времетраењето на болеста')
plt.legend()

plt.show()
```

### 7. How do factors influence the outcome whether a patient will have (or not) diabetic retinopathy? Analyse whether there is a correlation or not between diabetic retinopathy and other data in the set. Do this by separating the patients into 2 groups and analyzing the characteristics of the groups with respect to various information. Plot various data points if possible.

```python
# Креирање на променливи за понатамошна анализа

positive = df_filled[df_filled['Дијабетична ретинопатија'] == 1]
positive_mean = positive.drop(['ID', 'Тип на третман', 'Дијабетична ретинопатија'], axis=1).mean()

negative = df_filled[df_filled['Дијабетична ретинопатија'] == 0]
negative_mean = negative.drop(['ID', 'Тип на третман', 'Дијабетична ретинопатија'], axis=1).mean()

positive_count = positive['Дијабетична ретинопатија'].value_counts()
negative_count = negative['Дијабетична ретинопатија'].value_counts()

positive_max = positive.max()
positive_min = positive.min()

negative_max = negative.max()
negative_min = negative.min()


# Принтање на основен извештај

print('Просек на вредностите на пациентите со дијабетесна ретинопатија:')
print(positive_mean)
print('\n')
print('Просек на вредностите на пациентите без дијабетесна ретинопатија:')
print(negative_mean)
print('\n')

print('Број на позитивни пациенти:')
print(positive_count)
print('Број на негативни пациенти:')
print(negative_count)
```

```python
# Визуелна споредба на позитивни и негативни пациенти на дијабетесна ретинопатија

plt.figure(figsize=(15, 5))
plt.bar(['Позитивни', 'Негативни'], 
        [positive_count[1], negative_count[0]], 
        color=['#FFC107', '#00BFFF'])
plt.title('Споредба на позитивни и негативни на дијабетесна ретинопатија')
plt.xlabel('')
plt.ylabel('Број на заболени')

# Adding count values to the plot
for i, count in enumerate([positive_count[1], negative_count[0]]):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```

```python
# Споредба на моментално ниво на глукоза

plt.figure(figsize=(10, 6))
plt.boxplot([positive['Моментална крвна глукоза'],
             negative['Моментална крвна глукоза']],
            labels=['Позитивни на ретинопатија', 
                    'Негативни на ретинопатија'])

plt.xlabel('Дијабетесна ретинопатија')
plt.ylabel('Моментално ниво на глукоза')
plt.title('Споредба на моментално ниво на глукоза')
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# Моментално ниво на глукоза и HbA1C во последните 3 месеци

plt.figure(figsize=(10, 6))
plt.scatter(positive['Моментална крвна глукоза'], 
            positive['Ниво на HbA1C во последните 3 месеци'], 
            label='Позитивни на ретинопатија')
plt.scatter(negative['Моментална крвна глукоза'], 
            negative['Ниво на HbA1C во последните 3 месеци'], 
            label='Негативни на ретинопатија')
plt.xlabel('Моментално ниво крвна глукоза')
plt.ylabel('Ниво на HbA1C во последните 3 месеци')
plt.title('Моментално ниво на глукоза и HbA1C во последните 3 месеци')
plt.legend()
plt.show()
```

```python
# Споредба на максимални и минимални вредности

variables = ['Времетраење на дијабетес (во год.)', 
             'Ниво на HbA1C во последните 3 месеци', 
             'Тип на третман', 
             'Моментална крвна глукоза']

# Creating a dataframe for plotting
extreme_data = pd.DataFrame({
    'Positive_Max': [positive_max[variable] for variable in variables],
    'Positive_Min': [positive_min[variable] for variable in variables],
    'Negative_Max': [negative_max[variable] for variable in variables],
    'Negative_Min': [negative_min[variable] for variable in variables]
})

fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.2

positions = np.arange(len(variables))

ax.bar(positions - bar_width, extreme_data['Positive_Max'], width=bar_width, label='Positive_Max', color='#FFC107')
ax.bar(positions, extreme_data['Positive_Min'], width=bar_width, label='Positive_Min', color='#FFA000')
ax.bar(positions + bar_width, extreme_data['Negative_Max'], width=bar_width, label='Negative_Max', color='#00BFFF')
ax.bar(positions + 2 * bar_width, extreme_data['Negative_Min'], width=bar_width, label='Negative_Min', color='#0066CC')

for i, column in enumerate(extreme_data.columns):
    for j, value in enumerate(extreme_data[column]):
        ax.text(j + i * bar_width - bar_width/2, 
                value, 
                f'{value:.2f}', 
                ha='center', 
                va='bottom', 
                fontsize=10)

ax.set_xticks(positions)
ax.set_xticklabels(variables, rotation=45)

plt.title('Споредба на максимални и минимални вредности - Дијабетесна ретинопатија')
plt.xlabel(['Времетраење на дијабетес (во год.)', 
            'Ниво на HbA1C во последните 3 месеци', 
            'Тип на третман', 
            'Моментална крвна гликоза'])
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()
```
